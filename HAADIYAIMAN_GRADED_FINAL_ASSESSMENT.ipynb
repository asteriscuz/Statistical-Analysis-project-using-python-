```python
## 16- how many movies are in hindi? how many hindi movies are there after the year 2000?
```

```python
df.loc[df['language']=='Hindi'].shape[0]
```

```python
df.loc[(df['language']=='Hindi') & (df['year']>2000)].shape[0]
```

```python
## 17- group the movies by language and find out the avg budget of the country egypt
```

```python
d1=df.groupby(['language','country']).head()
d1
```

```python
d1.loc[d1['country']=='Egypt','budget'].mean()
```

```python
## 18- how many movies are from 1970s and imbd scores greator than 8?
```

```python
df.loc[(df['year'].between(1970,1979,inclusive=True)) & (df['imdb_score']>8)].shape[0]
```

```python
## 21- display the movies with imbd score with highest to lowest to find the avg imbd score
```

```python
d1=df.sort_values(by='imdb_score',ascending=False)
d1
```

```python
d1.agg({'imdb_score':'mean'})
```
